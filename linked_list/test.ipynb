> *LinkedList e DoublyLinkedList*

Imports

```python
import random
from linked_list import LinkedList
from doubly_linked_list import DoublyLinkedList
```

Instanciando LinkedList

```python
list = LinkedList()
print(list)
```

Método `push_back()`

```python
list.push_back(1)
list.push_back(2)
list.push_back(3)
list.push_back(4)
list.push_back(5)

print(list)
```

Método `push_front()`

```python
list.push_front(1)
list.push_front(2)
list.push_front(3)
list.push_front(4)
list.push_front(5)

print(list)
```

Método `pop_back()`

```python
list.pop_back()
print(list)
```

Método `pop_front()`

```python
list.pop_front()
print(list)
```

Método `insert_at()`

```python
list.insert_at(1, 1.5)
list.insert_at(3, 2.5)
list.insert_at(7, 6)

print(list)
```

Método `remove_at()`

```python
list.remove_at(1)
list.remove_at(3)
list.remove_at(5)

print(list)
```

Método `sort()`

```python
for _ in range(20):
    list.push_back(random.randint(0,100))

print("Lista desordenada:", list)

list.sort()

print("Lista ordenada:", list)
```

Método `copy()`

```python
list_copy = list.copy()

print(list_copy)
```

Método `concatenate()`

```python
new_list = LinkedList()
new_list.push_back(1)
new_list.push_back(2)
new_list.push_back(3)

print(new_list)

list.concatenate(new_list)

print(list)
```
