Método `sort()`

```python
for _ in range(20):
    list.push_back(random.randint(0,100))

print("Lista desordenada:", list)

list.sort()

print("Lista ordenada:", list)
```

Instanciando DoublyLinkedList

```python
list = DoublyLinkedList()
print(list)
```

Método `push_back()`

```python
list.insert(5)
list.insert(3)
list.insert(4)
list.insert(2)
list.insert(1)

print(list)
```

Método `remove_at()`

```python
list.remove_at(0)
list.remove_at(3)

print(list)
```

Método `pop_back()`

```python
for _ in range(len(list)):
    print(list.pop_back())
```

Método `search()`

```python
for _ in range(20):
    list.insert(random.randint(0, 100))

print(list)
pos = random.randint(0, len(list))
item = list.get_at(pos)

print(f"Item da posição {pos}:", item)
print(f"Posição da primeira ocorrência encontrada do item {item}:", list.search(item))

```
