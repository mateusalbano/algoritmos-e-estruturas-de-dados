Método `push_back()`

```python
deque.push_back(1)
deque.push_back(2)
deque.push_back(3)
deque.push_back(4)
deque.push_back(5)

print(deque)
```

Método `push_front()`

```python
deque.push_front(1)
deque.push_front(2)
deque.push_front(3)
deque.push_front(4)
deque.push_front(5)

print(deque)
```

Método `pop_back()`

```python
deque.pop_back()
print(deque)
```

```python
deque.pop_front()
print(deque)
```
