Imports

```python
from graph import Graph
```

Instanciando Grafo e Nós

```python
graph = Graph()
```

Método `add_node()`

```python
graph.add_node("A")
graph.add_node("B")
graph.add_node("C")
graph.add_node("D")

print(graph)
```

Método `add_edge()`

```python
graph.add_edge("A", "B", 1)
graph.add_edge("A", "C", 4)
graph.add_edge("B", "C", 2)
graph.add_edge("B", "D", 5)
graph.add_edge("C", "D", 1)
print(graph)
```

Método `pop_node()`

```python
popped_node = graph.pop_node("C")
print(f"Nó removido: {popped_node}")
print(graph)

graph.add_node("C")
graph.add_edge("A", "C", 4)
graph.add_edge("B", "C", 2)
graph.add_edge("C", "D", 1)

print()
print(graph)
```

Método `has_cycle()`

```python
graph.remove_edge("B", "C")
graph.remove_edge("A", "B")
graph.add_edge("D", "C")
print(graph)
print("has_cycle():", graph.has_cycle())
print()
graph.add_edge("B", "C", 2)
graph.add_edge("B", "A", 4)
graph.add_edge("C", "B", 3)
print(graph)
print("has_cycle():", graph.has_cycle())
```

Método `has_path()`

```python
graph.add_node("E")
print(graph)
print("has_path('E', 'A'):", graph.has_path("E", "A"))
print()

graph.add_edge("E", "D", 2)
print(graph)
print("has_path('E', 'A'):", graph.has_path("E", "A"))
```
