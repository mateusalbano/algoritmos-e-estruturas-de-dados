Imports

```python
import sys
from pathlib import Path

project_root = Path("..").resolve()

if str(project_root) not in sys.path:
    sys.path.insert(0, str(project_root))


from graph.graph import Graph, Node
```

Instanciando Grafo e Nós

```python
graph = Graph()
node_a = Node('A')
node_b = Node('B')
node_c = Node('C')
node_d = Node('D')
```

Método `add_node()`

```python
graph.add_node(node_a)
graph.add_node(node_b)
graph.add_node(node_c)
graph.add_node(node_d)

print(graph)
```

Método `add_edge()`

```python
graph.add_edge(node_a, node_b, 1)
graph.add_edge(node_a, node_c, 4)
graph.add_edge(node_b, node_c, 2)
graph.add_edge(node_b, node_d, 5)
graph.add_edge(node_c, node_d, 1)

print(graph)
```

Método `pop_node()`

```python
popped_node = graph.pop_node(node_c)
print(f"Nó removido: {popped_node}")
print(graph)

graph.add_node(node_c)

graph.add_edge(node_a, node_c, 4)
graph.add_edge(node_b, node_c, 2)
graph.add_edge(node_c, node_d, 1)

print()
print(graph)
```

Método `has_cycle()`

```python
graph.remove_edge(node_b, node_c)
graph.remove_edge(node_a, node_b)
print(graph)
print("has_cycle():", graph.has_cycle())
print()
graph.add_edge(node_b, node_c, 2)
graph.add_edge(node_b, node_a, 4)
graph.add_edge(node_c, node_b, 3)
print(graph)
print("has_cycle():", graph.has_cycle())
```
