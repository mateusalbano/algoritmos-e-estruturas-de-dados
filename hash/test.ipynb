Imports

```python
from hash_map import HashMap
```

Instanciando Hash Map

```python
map = HashMap()
```

Método `put()`

```python
map.put("Mateus", 21)
map.put("João", 20)
map.put("Pedro", 19)

print(map)
```

Método `get()`

```python
print(map.get("Mateus"))
print(map.get("João"))
print(map.get("Pedro"))
```

Método `remove()`

```python
map.remove("João")
print(map)

print(map.get("João"))
```
