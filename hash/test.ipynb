> *HashSet e HashMap* 

Imports

```python
from hash_set import HashSet
from hash_map import HashMap
```

Instanciando Hash Set

```python
set = HashSet()
```

Método `put()`

```python
set.put("Mateus")
set.put("João")
set.put("Pedro")

print(set)
```

Método `contains()`

```python
print(set.contains("Mateus"))
print(set.contains("João"))
print(set.contains("Pedro"))

print(set.contains("Felipe"))
```

Método `remove()`

```python
set.remove("João")
print(set)

print(set.contains("João"))
```

Instanciando Hash Map

```python
map = HashMap()
```

Método `put()`

```python
map.put("Mateus", 21)
map.put("João", 20)
map.put("Pedro", 19)

print(map)
```

Método `get()`

```python
print(map.get("Mateus"))
print(map.get("João"))
print(map.get("Pedro"))
```

Método `remove()`

```python
map.remove("João")
print(map)

print(map.get("João"))
```
