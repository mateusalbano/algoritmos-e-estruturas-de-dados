```python
import random
import sys
import os

if 'queue' in sys.modules:
    del sys.modules['queue']

import os

caminho_atual = os.getcwd() + "\\.."
sys.path.insert(0, caminho_atual)

from queue.queue import Queue
from queue.heap import Heap
```

Instanciando Queue

```python
queue = Queue()
print(queue)
```

Método `enqueue()`

```python
queue.enqueue(1)
queue.enqueue(2)
queue.enqueue(3)
queue.enqueue(4)
queue.enqueue(5)

print(queue)
```

Método `dequeue()`

```python
for _ in range(len(queue)):
    print(queue.dequeue())
```

Instanciando Heap

```python
queue = Heap()
```

Método `enqueue()`

```python
for _ in range(15):
    queue.enqueue(random.randint(0,100))
print(queue)
```
