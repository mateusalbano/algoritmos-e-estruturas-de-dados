Imports

```python
import random
from array_list import ArrayList
```

Instanciando ArrayList

```python
list = ArrayList()
print(list)
```

Método `push_back()`

```python
list.push_back(1)
list.push_back(2)
list.push_back(3)
list.push_back(4)
list.push_back(5)

print(list)
```

Método `insert_at()`

```python
list.insert_at(1, 1.5)
list.insert_at(3, 2.5)
list.insert_at(7, 6)

print(list)
```

Método `remove_at()`

```python
list.remove_at(1)
list.remove_at(3)
list.remove_at(5)

print(list)
```

Método `pop_back()`

```python
for _ in range(len(list)):
    print(list.pop_back())
```

Método `sort()`

```python
for _ in range(20):
    list.push_back(random.randint(0,100))

print("Lista desordenada:", list)

list.sort()

print("Lista ordenada:", list)
```

Método `copy()`

```python
list_copy = list.copy()
print(list_copy)
```

Método `concatenate()`

```python
new_list = ArrayList()
new_list.push_back(1)
new_list.push_back(2)
new_list.push_back(3)

print(new_list)

list.concatenate(new_list)
print(list)
```
